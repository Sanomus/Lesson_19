```python
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset("penguins")

penguins.head()
penguins.dropna()
```

```python
penguins.describe()
```

```python
fig, axs = plt.subplots(ncols=1)
fig.set_size_inches(16, 5)
#mass distribution of all penguins
sns.histplot(penguins['body_mass_g'], binwidth=10, ax=axs)
```

```python
#we can see mass and flipper length of deferent species
sns.scatterplot(x = 'body_mass_g', y = 'flipper_length_mm',  data = penguins, hue = 'species')
```

```python
#we can see mass and flepper length on deferent island
sns.scatterplot(x = 'body_mass_g', y = 'flipper_length_mm',  data = penguins, hue = 'island')
```

```python
#better represent destribution  of mass in species between male an female
sns.scatterplot(x = 'species', y = 'body_mass_g', data = penguins, hue = 'sex')
```

```python
#better represent destribution  of mass in species between male an female
sns.catplot(x = 'species', y = 'body_mass_g', kind='swarm', data = penguins, hue = 'sex')
```

```python
#in swarm we ccn see better
sns.catplot(x = 'species', y = 'body_mass_g', kind='strip', data = penguins, hue = 'sex')
```

```python

_ = sns.catplot(x="species", y="flipper_length_mm", hue="island", kind="box", orient="v", data=penguins)
```

```python
sns.catplot(x="island", y="flipper_length_mm", hue="species", kind="violin", orient="v", data=penguins)
```

```python
#dependecies mass with flipper length
_ = sns.regplot(x="flipper_length_mm", y="body_mass_g", data=penguins)
```
